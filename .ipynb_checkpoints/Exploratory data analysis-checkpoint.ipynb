### Building Permits Data Cleaning
Get the ADU related building permits and only keep the latest record

```python
ADU_permits = pd.read_csv(data + 'Building_Permits_LA_ADU_filter.csv')
```

```python
ADU_permits.shape
```

```python
ADU_permits.head()
```

```python
# ADU_permits.columns
```

```python
# ADU_permits.isnull().sum()
```

```python
ADU_build_permits = ADU_permits[ADU_permits['Permit Type'].isin(['Bldg-Addition', 'Bldg-New',
                                                                 'Bldg-Alter/Repair', 'Nonbldg-New'])]
```

```python
ADU_permits_clean = ADU_build_permits[['Assessor Book', 'Assessor Page', 'Assessor Parcel', 
                                       'Work Description', 'Issue Date',
                                       'Valuation', 'Floor Area-L.A. Zoning Code Definition',
                                       '# of Residential Dwelling Units', '# of Stories', 'Zone']].copy()
```

```python
ADU_permits_clean['Assessor Book'] = ADU_permits_clean['Assessor Book'].apply(lambda x: '{0:0>4}'.format(x))
ADU_permits_clean['Assessor Page'] = ADU_permits_clean['Assessor Page'].apply(lambda x: '{0:0>3}'.format(x))
ADU_permits_clean['Assessor Parcel'] = ADU_permits_clean['Assessor Parcel'].apply(lambda x: '{0:0>3}'.format(x))
```

```python
ADU_permits_clean['APN_number'] = ADU_permits_clean['Assessor Book'].astype(str) + \
                                  ADU_permits_clean['Assessor Page'].astype(str) + \
                                  ADU_permits_clean['Assessor Parcel'].astype(str)
```

```python
ADU_permits_clean = ADU_permits_clean.drop(['Assessor Book','Assessor Page','Assessor Parcel'], axis = 1)
```

```python
ADU_permits_clean.dtypes
```

```python
ADU_permits_clean['Issue Date'] =  pd.to_datetime(ADU_permits_clean['Issue Date'], format='%m/%d/%Y')
```

```python
ADU_permits_clean.shape
```

```python
ADU_permits_clean.head(10)
```

```python
latest_permit = ADU_permits_clean.sort_values('Issue Date').groupby('APN_number').tail(1)
```

```python
latest_permit.shape
```

```python
# ADU_permits_clean.APN_number.nunique() ==  ADU_permits_clean.shape[0]
```
